```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
titanic=pd.read_csv('/content/train.csv')
titanic.head()
```

```python
flight=sns.load_dataset('flights')
flight.head()
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

1. scatterplot(numerical-numerical)

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker',size='size')
```

2. barplot(numerical-categorical)

```python
titanic.head()
```

```python
sns.barplot(x='Pclass',y='Age',data=titanic)
```

```python
sns.barplot(x='Pclass',y='Fare',data=titanic,hue='Sex')
```

3. boxplot(numerical-categorical)

```python
sns.boxplot(x='Sex',y='Age',data=titanic,hue='Survived')
```

5. histplot(numerical-categorical)

```python
sns.histplot(x='Age', data=titanic,kde=True,hue='Survived')
```

```python
sns.kdeplot(data=titanic, x='Age', hue='Survived')
```

5. heatmap(categorical-categorical)

```python
titanic.head()
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
#in percentage terms how many people survived in each pclass
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
#in percentage terms how many people survived in each Embarked
titanic.groupby('Embarked')['Survived'].mean()*100
```

6. clustermap(categorical-categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

7. pairplot

```python
#plots scatterplot with every numerical column in dataset
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

8. lineplot(numerical-numerical)

```python
flight.head()
```

```python
flight.groupby('year')['passengers'].sum().reset_index()
```

```python
new = flight.groupby('year')['passengers'].sum().reset_index()
sns.lineplot(x='year', y='passengers', data=new)
```

```python
flight.pivot_table(index='year', columns='month', values='passengers')
```

```python
sns.heatmap(flight.pivot_table(index='year', columns='month', values='passengers'))
```

```python
sns.clustermap(flight.pivot_table(index='year', columns='month', values='passengers'))
```
